### TOKENITZACIÓ

```python
# La tokenització, en l'àmbit del processament del llenguatge natural (NLP) i l'aprenentatge automàtic, 
# es refereix al procés de convertir una seqüència de text en parts més petites, conegudes com a tokens. 
# Aquestes fitxes poden ser tan petites com caràcters o tan llargues com paraules.
# nltk.tokenize és un paquet que ens permet tokenitzar textos. Forma part del Natural Language Toolkit.
!pip install nltk
import nltk.data
from nltk.tokenize import word_tokenize, RegexpTokenizer, punkt
```

```python
# Forma 1 - Dona error (?) -> La comentem.
# word_tokenized = word_tokenize(Poema, language="english", preserve_line=False))
# word_to# kenized
```

```python
# Forma 2 - Funciona 
rt = RegexpTokenizer(r'\w+')
regex_tokenized = rt.tokenize(Poema)
regex_tokenized
```

```python
# Podem pintar el resultat
print(regex_tokenized)
```

```python
# Podem analitzar la frecuencia en que apareixen determinades paraules
# Fem servir el mòdul nltk probability i el mètode FreqDist
from nltk.probability import FreqDist
```

```python
# fd_word = FreqDist(word_tokenized)
fd_regex = FreqDist(regex_tokenized)
```

```python
# Ens genera un diccionari on la clau es la paraula i el valor el nombre de cops que apareix en el text analitzat. 
fd_regex
```

```python
# Graficament, ens mostra ordenat de major a menor nombre d'aparicions en el text cada paraula
fd_regex.plot(20);
```

```python
# Eliminació de caràcters especials i signes de puntuació
wt_without_signes = [word for word in regex_tokenized if word.isalpha()]
wt_without_signes
```

```python
print (wt_without_signes)
```

```python
# Normalització
# Eliminem tots els signes de puntuació i caracters espècials. En idioma anglès hi ha menys que en altres idiomes.
wt_lower = [word.lower() for word in wt_without_signes]
rt_lower = [word.lower() for word in regex_tokenized]
```

```python
print(wt_lower)
```

```python
print(rt_lower)
```

### Exercici 2: Treu les stopwords i realitza stemming al teu conjunt de dades.

```python
# Stopwords
# En el processament del llenguatge natural (NLP), els stopwords es filtren amb freqüència per millorar l'anàlisi de text 
# i l'eficiència computacional. Eliminar paraules clau pot millorar la precisió i la rellevància de les tasques de PLN cridant 
# l'atenció sobre les paraules més importants, o paraules de contingut.
import nltk
from nltk.corpus import stopwords
```

```python
# Mostrem els stopwords (paraules ineficaces) en un text en anglès i mostrem la impressió i la llista ordenada alfabèticament.
# Certes paraules, com “the,” “and”, i “is”, en anglès, són ineficaces per comunicar informació important. 
# L'objectiu d'eliminar paraules que afegeixen poc o res a la comprensió del text és accelerar el processament del text, 
# tot i que la llista de paraules clau pot diferir.

stops = set(stopwords.words('english'))
print(stops)
stops
```

```python
# Eliminem les paraules 'stopwords' del text en anglès original que apareixen a la llista anterior.
wt_stops = [word for word in wt_lower if not word in stops]
rt_stops = [word for word in rt_lower if not word in stops]
```

```python
# Mostrem les paraules restants en una impressió del resultat
print(wt_stops)
```

```python
print(rt_stops)
```

```python
stops.add('na')
```

```python
wt_stops = [word for word in wt_lower if not word in stops]
```

```python
print(wt_stops)
```

```python
# Stemming
# L'algorisme de Porter és un algorisme per a stemming que assegura que la morfologia de les paraules no penalitzi 
# la freqüència d'aquestes. És a dir, un morfema pot estar unit per modificadors de gènere, número o persona i es considerarà com un únic terme.
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
wt_stemmer = [ps.stem(word) for word in wt_stops]
```

```python
print(wt_stemmer)
```

```python
# Lemmatització
# La lemmatització és una tècnica fonamental de preprocessament de textos àmpliament aplicada en el processament del llenguatge natural (NLP)
# i l'aprenentatge automàtic. La lemmatització busca destil·lar paraules a les seves formes fundacionals. 
# En aquest refinament lingüístic, la paraula base resultant es coneix com a “lemma.”
from nltk.stem import WordNetLemmatizer
```

```python
# WordNet és una base de dades lèxica de l'Idioma inglés1​ que agrupa paraules en anglès en conjunts de sinònims anomenats synsets, 
# proporcionant definicions curtes i generals i emmagatzemant les relacions semàntiques entre els conjunts de sinònims.
nltk.download('wordnet') 
```

```python
# En aplicar-ho dona error -> Ho deixo comentat.
wnl = WordNetLemmatizer()
rt_lemmatize = [wnl.lemmatize(word) for word in rt_stops]
```

```python
rt_lemmatize
```

```python
# Visualitzacions
fd_wt_stem = FreqDist(wt_stemmer)
fd_rt_lem = FreqDist(rt_lemmatize)
```

```python
fd_wt_stem.plot(20);
```

```python
fd_rt_lem.plot(20);
```

### Exercici 3: Realitza sentiment analysis al teu conjunt de dades.

#### TextBlob

```python
# TextBlob
# TextBlob és una biblioteca de Python per processar dades textuals. 
# Proporciona una API senzilla per submergir-se en tasques comunes de processament del llenguatge natural (NLP), 
# com ara etiquetat de part de veu, extracció de frases nominals, anàlisi de sentiment, classificació i més.
!pip install textblob
from textblob import TextBlob
```

```python
# Apliquem la API a la llista wt_stemmer.
tb_wt = TextBlob('wt_stemmer')
```

```python
# Subjectivitat
# La subjectivitat quantifica la quantitat d'opinió personal i informació objectiva continguda en el text. 
# L'elevada subjectivitat significa que el text conté opinió personal en lloc d'informació objectiva. 
# TextBlob té un paràmetre més — intensitat. TextBlob calcula la subjectivitat mirant la “intensitat”. 
# La intensitat determina si una paraula modifica la paraula següent. 
# Per a l'anglès, els adverbis s'utilitzen com a modificadors («very good»).
tb_wt = TextBlob(' '.join(wt_stemmer))
tb_wt.subjectivity
```

```python
# Polaritat
# La polaritat es troba entre [-1,1], -1 defineix un sentiment negatiu i 1 defineix un sentiment positiu. 
# Les paraules de negació reverteixen la polaritat. TextBlob té etiquetes semàntiques que ajuden a l'anàlisi de gra fi. 
# Per exemple — emoticones, marca d'exclamació, emojis, etc.
tb_wt.polarity
```

```python
# VADER
# És un lèxic i un instrument d'anàlisi de sentiments basat en regles que és explícitament sensible als supòsits comunicats en mitjans web. 
# VADER utilitza una barreja de ressaltats lèxics (per exemple, paraules) que estan, en la seva major part, 
# marcats per la seva direcció semàntica com un o altre positiu o negatiu.
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
# Calcula les etiquetes de sentiment
nltk.download('vader_lexicon') 
```

```python
# Ens valora els percentatges de negativitat, positivitat, neutralitat així com un factor compost
sia = SentimentIntensityAnalyzer()
sia.polarity_scores(' '.join(wt_stemmer))
```
